```python
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.arange(10.3, 663.2, 2.53)
```

```python
x
```

```python
y = np.random.normal(x, 80)
```

```python
plt.scatter(x, y)
```

```python
df = pd.DataFrame({
    'Masa (kg)' : x,
    'Cijena transporta (HRK)' : y
})
```

```python
df = df.loc[(df['Cijena transporta (HRK)'] > 5) & (df['Cijena transporta (HRK)'] < 1000)]
```

```python
df['Cijena transporta (HRK)'] = df['Cijena transporta (HRK)'] * 11.779
```

```python
plt.scatter(df['Masa (kg)'], df['Cijena transporta (HRK)'])
```

```python
df.to_excel('podaci.xlsx')
```
