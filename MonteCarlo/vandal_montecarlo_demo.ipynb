vandal.MonteCarlo demonstration

```python
import vandal # vandal 3.8.1 was used in this example.
import pandas as pd # importing pandas to create a DataFrame.
```

```python
vandal.__version__ # version used.
```

```python
MC = vandal.MonteCarlo({
    'Example_data' : [38,37,38,36,36,35,36,34,33,34,35], # example list of values. (ref_col = 0)
    'Example_data2' : [142, 149, 153, 154, 150, 149, 149, 151, 152, 150, 146], # example list of values. (ref_col = 1)
    },
50, 
300, 
ref_col= 1,
ref_row = 6,
) # object defined as MC.
```

```python
MC.execute(filtered=False) # executes the Monte Carlo simulation.
```

```python
MC.graph() # creates a graph of the executed Monte Carlo simulation.
```

```python
MC.get_change() # returns the list of percentage changes throughout the executed simulation.
```

```python
print(MC.get_risk(15000)) # shows the risk evaluation on top of the number of simulations defined in the parentheses.
```

```python
MC.get_stats() # shows the statistics of the executed Monte Carlo simulation.
```

```python
MC.hist(graph_title='GitHub Repository Demo', x_title = 'Time period', y_title = 'Value') # histogram of values.
```

```python
MC.hist(x_title = 'Time period', y_title = 'Value', method = 'e') # histogram with Empirical rule applied, BEWARE: this option is currently being worked on so do not use it in any research!!!
```
