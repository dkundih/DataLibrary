```python
# unos biblioteka.
import kundih
import pandas as pd
import folium
```

```python
# unos lokacija i filtriranje.
lokacije = pd.read_excel(r'import_data/test_lok.xlsx')
lokacije.drop(columns = 'Unnamed: 0', inplace = True)
```

```python
# prikaz prvih deset lokacija.
lokacije.head(10)
```

```python
# inicijalni prikaz karte.
karta = folium.Map(location = [45.80,15.96], zoom_start=13)
karta
```

```python
# dodavanje markera na kartu.
for i in range(len(lokacije['Geo_širina'])):
    folium.Marker(location=[lokacije['Geo_dužina'][i], lokacije['Geo_širina'][i]]).add_to(karta)
karta
```

```python
# inicijalizacija objekta.
KSV = kundih.KSrednjeVrijednosti(datoteka = lokacije, broj_klastera = 3, koordinate = ['Geo_širina', 'Geo_dužina'])
```

```python
# podešavanje omjera (fitting).
KSV.podesi_omjer().head(10)
```

```python
# metoda koja kreira klastere nad podacima.
KSV.kreiraj_klastere()
```

```python
# metoda koja prikazuje lokacije centroida.
KSV.prikaz_centroida()
```

```python
# prikazivanje grafa s 3 klastera.
f = KSV.prikaz_grafa()
```
