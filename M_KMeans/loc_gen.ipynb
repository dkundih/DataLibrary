```python
import random
import pandas as pd
```

```python
x = [random.uniform(15.84, 16.13) for i in range(300)]
y = [random.uniform(45.75, 45.86) for i in range(300)]
```

```python
test_lok = pd.DataFrame({
    'Geo_širina' : x,
    'Geo_dužina' : y,
})
```

```python
import matplotlib.pyplot as plt
plt.scatter(x= test_lok['Geo_širina'], y = test_lok['Geo_dužina'])
```

```python
test_lok.to_excel('test_lok.xlsx')
```
