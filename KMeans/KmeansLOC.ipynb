```python
import folium
```

```python
import pandas as pd
df = pd.read_excel('Lokacije.xlsx')
```

```python
m = folium.Map(location = [46.39,16.43], zoom_start=14)
m
```

```python
koordinate1 = df['L2']
koordinate2 = df['L1']
```

```python
for i in range(len(koordinate1)):
    folium.Marker(location=[koordinate1[i], koordinate2[i]]).add_to(m)
```
