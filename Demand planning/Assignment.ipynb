```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r'C:\Users\kundi\Sphere\DataLibrary\Demand planning\import_data\data.xlsx')
```

```python
df
```

```python
df['CUMULATIVE'] = df['JAN'] + df['FEB'] + df['MAR'] + df['APR'] + df['MAY'] + df['JUN'] + df['JUL'] + df['AUG']
```

```python
df
```

```python
df['STOCK/DEMAND RATIO'] = df['QUANTITY'].astype('float64').divide(df['CUMULATIVE']).astype('float64')
```

```python
df
```

```python
plt.figure(figsize=(10,3))
plt.title('STOCK/DEMAND RATIO', size= 20)
plt.scatter(x = df['PRODUCT'], y = df['STOCK/DEMAND RATIO'])
plt.ticklabel_format(style='plain', axis='y')
plt.axhline(0, color = 'black')
plt.xticks(rotation=90)
plt.show()
```
