0:3 CLASSES FOR KERAS PROCESSING

```python
df['Mode'].replace(1,0,inplace=True)
df['Mode'].replace(2,1,inplace=True)
df['Mode'].replace(3,2,inplace=True)
```

TRAINING

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
```

```python
x = df.drop(columns = ['Mode'])
y = df['Mode']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

SETTING 3 CLASSES

```python
y_train = to_categorical(y_train, num_classes = 3) ## Convert to categorical
y_test = to_categorical(y_test, num_classes = 3) ## num_classes = 3 because there are 3 different modes of transport
```

```python
model = tf.keras.models.Sequential() ## Sequential is a linear stack of layers
```

```python
x_train.shape
```

```python
model.add(tf.keras.layers.Dense(256,input_shape = x_train.shape, activation='relu')) 
model.add(tf.keras.layers.Dense(256, activation = 'relu'))
model.add(tf.keras.layers.Dense(256, activation = 'relu'))
model.add(tf.keras.layers.Dense(256, activation = 'relu')) 
model.add(tf.keras.layers.Dense(256, activation = 'relu')) 
model.add(tf.keras.layers.Dense(3, activation='softmax')) ## Softmax is used to convert the output to probabilities, which is what we want for classification problems.
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs = 300, batch_size = 32)
```

EVAULATION

```python
model.evaluate(x_test, y_test)
```
